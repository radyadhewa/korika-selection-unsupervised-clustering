```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("/kaggle/input/pedass/Soal Final Pedas 2023.xlsx", sheet_name="Data_Capaian_SIPSN_KLHK")
```

```python
df_by_year = df.groupby("Tahun")
```

```python
df_by_year_list = [df_by_year.get_group(x) for x in df_by_year.groups]
```

```python
df_2020 = df_by_year_list[0]
df_2021 = df_by_year_list[1]
df_2022 = df_by_year_list[2]
```

```python
df_2022_by_provinsi = df_2020.groupby("Provinsi").sum()

not_features = ["Tahun", "Negara", "Kabupaten/Kota", "Timbulan Sampah Tahunan (ton/tahun) (A)"]
label = ["Provinsi"]

features = [c for c in df_2022_by_provinsi.columns if c not in not_features + label]
```

```python
features
```

```python
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN, KMeans
from sklearn.preprocessing import StandardScaler, QuantileTransformer
from sklearn.decomposition import PCA


scaler = StandardScaler().set_output(transform="pandas")

q_transformer = QuantileTransformer(
    n_quantiles=1000,
    output_distribution='normal',
    ignore_implicit_zeros=False,
    subsample=10000,
    random_state=None,
    copy=True
)

X = df_2022_by_provinsi[features]
X = scaler.fit_transform(X)
X = q_transformer.fit_transform(X)


# dbscan = DBSCAN(
#     eps=0.5,
#     min_samples=5,
#     metric='euclidean',
#     algorithm='auto',
#     leaf_size=30,
#     p=None,
#     n_jobs=None
# )


t_sne = TSNE(
    n_components=2,
    perplexity=30,
    early_exaggeration=20,
    init="random",
    n_iter=250,
    random_state=0,
)

pca = PCA(
    n_components=2,
    random_state=42,
)
```

```python
from sklearn.metrics import silhouette_score



# Define a range of values for n_clusters
n_clusters_range = range(2, 11)  # You can adjust the range as needed

# Initialize variables to store the best score and corresponding n_clusters
best_score = -1
best_n_clusters = -1

# Loop through different values of n_clusters
for n_clusters in n_clusters_range:
    kmeans = KMeans(n_clusters=n_clusters, n_init="auto")
    clusters = kmeans.fit_predict(X)
    
    # Calculate the silhouette score for the current n_clusters
    silhouette_avg = silhouette_score(X, clusters)
    print(f"n_clusters={n_clusters} : {silhouette_avg}")
    
    # Check if the current score is better than the previous best
    if silhouette_avg > best_score:
        best_score = silhouette_avg
        best_n_clusters = n_clusters

# Print the best n_clusters and silhouette score
print(f"Best number of clusters: {best_n_clusters}")
print(f"Best silhouette score: {best_score}")
```

```python
year = 2020
best_n_clusters = {}

for data in df_by_year_list:
    df_by_provinsi = data.groupby("Provinsi").sum()

    scaler = StandardScaler().set_output(transform="pandas")

    q_transformer = QuantileTransformer(
        n_quantiles=1000,
        output_distribution='normal',
        ignore_implicit_zeros=False,
        subsample=10000,
        random_state=None,
        copy=True
    )

    X = df_by_provinsi[features]
    X = scaler.fit_transform(X)
    X = q_transformer.fit_transform(X)

    silhouette_scores = []
    for n_clusters in range(2, 11):  
        kmeans = KMeans(n_clusters=n_clusters, n_init="auto", random_state=42, max_iter=5000)
        clusters = kmeans.fit_predict(X)
        silhouette_avg = silhouette_score(X, clusters)
        silhouette_scores.append((n_clusters, silhouette_avg))

    best_n_clusters = max(silhouette_scores, key=lambda x: x[1])[0]

    X_embedding = pca.fit_transform(X)
    
    best_kmeans = KMeans(n_clusters=best_n_clusters, n_init="auto", random_state=42, max_iter=5000)

    df_output = pd.DataFrame(X_embedding, index=df_by_provinsi.index, columns=['Component 1', 'Component 2'])
    df_output["Cluster"] = best_kmeans.fit_predict(X)
    df_output["Tahun"] = year

    # Calculate the average for each column
    column_averages = np.mean(best_kmeans.cluster_centers_, axis=1)

    print("Average Centroid for each cluster center:")
    for i, average in enumerate(column_averages):
        print(f"{i}: {average}")
    
    print(f"Year: {year}, Best n_clusters: {best_n_clusters}")
    print(df_output)

    df_output.to_csv(f"clustering_capaian_sipsn_{year}.csv", index=True)
    year += 1
```

```python
results = [
    "/kaggle/working/clustering_capaian_sipsn_2020.csv",
    "/kaggle/working/clustering_capaian_sipsn_2021.csv",
    "/kaggle/working/clustering_capaian_sipsn_2022.csv"
]

# final_df = pd.concat((pd.read_csv(f) for f in results), ignore_index=True)
```

```python
df_2020, df_2021, df_2022 = [pd.read_csv(f) for f in results]
```

```python
cluster_mapping_2020 = {
    0: "Bad",
    1: "Good"
}

cluster_mapping_2021 = {
    0: "Bad",
    1: "Poor",
    2: "Good",
    3: "Best",
    4: "Worst",
}

cluster_mapping_2022 = {
    0: "Good",
    1: "Bad"
}

df_2020["Cluster"] = df_2020["Cluster"].map(cluster_mapping_2020)
df_2021["Cluster"] = df_2021["Cluster"].map(cluster_mapping_2021)
df_2022["Cluster"] = df_2022["Cluster"].map(cluster_mapping_2022)
```

```python
final_df = pd.concat([df_2020, df_2021, df_2022], ignore_index=True)
```

```python
final_df.to_csv("hasil_clustering_capaian.csv", index=False)
```
